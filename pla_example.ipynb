```python
import matplotlib.pyplot as plt
import numpy as np
origin = [0], [0]
import random
```

```python
#創立dataset
def linearfunction(x1,x2):
    data=2*x1-3*x2+1
    return data
a=np.random.random((20,2))*4-1
```

```python
ylist=[]
for i in range(20):
    y=linearfunction(a[i,0],a[i,1])
    if y>0:
        ylist.append(1)
    else:
        ylist.append(-1)
    
```

```python
data=np.hstack((np.ones((20,1)),a))
```

```python
dataset=[]
for i in range(20):
    dataset.append(((data[i,0],data[i,1],data[i,2]),ylist[i]))
```

```python
#判斷有沒有分類錯誤，並列印錯誤率

def check_error(w, dataset):
    result = (0,0)
    error = 0
    resultx=(0,0,0)
    results=0
    for x, s in dataset:
        x = np.array(x)
        if int(np.sign(w.T.dot(x))) != s:
            resultx,results =  x, s
            error += 1
            
    resulte=error
    print("error=%s/%s" % (error, len(dataset)))
    
    ps = [v[0] for v in dataset]
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    #dataset前半後半已經分割好 直接畫就是
    for x, s in dataset:
        if s==1:
            ax1.scatter([x[1]], [x[2]], s=10, c='b', marker="o", label='O')
        else:
            ax1.scatter([x[1]], [x[2]], s=10, c='r', marker="x", label='X')
    plt.quiver(*origin, resultx[1], resultx[2], angles='xy', scale_units='xy', scale=1,color='red')
    plt.quiver(*origin, w[1], w[2], angles='xy', scale_units='xy', scale=1)
    l = np.linspace(-2,2)
    a,b = -w[1]/w[2], -w[0]/w[2]
    ax1.plot(l, a*l + b, 'b-')
    plt.legend(loc='upper left');
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axis('equal')
    plt.show()
    result=(resultx,results,resulte)
    
    
    return result

#PLA演算法實作

def pla(dataset):
    w = np.array([1,0.,0.])
    print(w)
    x=1
    while x is not None :
        x, s, e = check_error(w, dataset)
        if e==0 :
            break
        fig = plt.figure()
        ax1 = fig.add_subplot(111)
        plt.quiver(*origin, x[1], x[2], angles='xy', scale_units='xy', scale=1,color='red')
        plt.quiver(*origin, w[1], w[2], angles='xy', scale_units='xy', scale=1)
        
        w[1] += s * x[1]
        w[2] += s * x[2]
        plt.quiver(*origin, w[1], w[2], angles='xy', scale_units='xy', scale=1,color="blue")
        
        plt.xlim(-2, 2)
        plt.ylim(-2, 2)
        plt.axis('equal')
        plt.show()
        print(w)
        
    return w


#執行

w = pla(dataset)

#畫圖

```
